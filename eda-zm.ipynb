```python
# tweet_zm_df[tweet_zm_df['date'].dt.year == 2020].shape

# getting tweets within 2020 and 2022
tweet_zm_2019_df = tweet_zm_df[tweet_zm_df['date'].dt.year == 2019]
tweet_zm_2020_df = tweet_zm_df[tweet_zm_df['date'].dt.year == 2020]
tweet_zm_2021_df = tweet_zm_df[tweet_zm_df['date'].dt.year == 2021]
tweet_zm_2022_df = tweet_zm_df[tweet_zm_df['date'].dt.year == 2022]



print('tweets 2019 - zm prices up due to covid')
print(tweet_zm_2019_df.shape)

print('tweets 2020 - zm prices up due to covid')
print(tweet_zm_2020_df.shape)

print('tweets 2021 - zm prices stable i think, but still high')
print(tweet_zm_2021_df.shape)

print('tweets 2022 - price goes down')
print(tweet_zm_2022_df.shape)

print ('notice the number of tweets per year is going down. But take note that \
     2022 is not full and 2019 is not full either as zm went public in april 2019')

print('next see each month tweets')

# once we get each month, test how many likes, retweets etc... and see if there a pattern.
# average price per month and how many tweets per that month.
# later we can check 1 week of sentiment to each 1 day of price.
```

```python
# Trying Grouper

# group_by_month_tweet_df = tweet_zm_df.groupby(pd.Grouper(freq='M'))

# group_by_month_tweet_df.sum()


# simple method
group_by_month_tweet_df = tweet_zm_df.groupby([(tweet_zm_df.date.dt.year), (tweet_zm_df.date.dt.month)]).sum()


group_by_month_tweet_df
```
